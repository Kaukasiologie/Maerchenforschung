```python
import os
from xml.dom import minidom

# ---------------------------------------------
# Einstellungen
# ---------------------------------------------

# Ordner, der durchsucht werden soll
SOURCE_DIR = r"Textkorpus"

# Falls die formatierten XMLs woanders gespeichert werden sollen
# TARGET_DIR = r"/Pfad/zum/Zielordner"
# Wenn du überschreiben möchtest, setze TARGET_DIR = SOURCE_DIR
TARGET_DIR = SOURCE_DIR

# Liste von Dateiendungen, die als XML behandelt werden dürfen
XML_EXTENSIONS = [".xml"]


# ---------------------------------------------
# Hilfsfunktion: XML formatieren
# ---------------------------------------------

def format_xml_file(input_path, output_path):
    try:
        # Datei lesen
        with open(input_path, "r", encoding="utf-8") as f:
            content = f.read()

        # XML parsen und formatieren
        parsed = minidom.parseString(content)
        pretty = parsed.toprettyxml(indent="  ", encoding="utf-8").decode("utf-8")

        # 🧹 Leerzeilen entfernen
        pretty = "\n".join([line for line in pretty.split("\n") if line.strip()])

        # Datei schreiben
        with open(output_path, "w", encoding="utf-8") as f:
            f.write(pretty)

        return True

    except Exception as e:
        print(f"⚠️ Fehler in Datei {input_path}: {e}")
        return False


# ---------------------------------------------
# Hauptlogik: Ordner rekursiv durchsuchen
# ---------------------------------------------
count = 0

for root, dirs, files in os.walk(SOURCE_DIR):
    for file in files:
        if any(file.lower().endswith(ext) for ext in XML_EXTENSIONS):
            input_path = os.path.join(root, file)

            # Zielpfad erzeugen
            if TARGET_DIR == SOURCE_DIR:
                output_path = input_path
            else:
                rel = os.path.relpath(root, SOURCE_DIR)
                out_dir = os.path.join(TARGET_DIR, rel)
                os.makedirs(out_dir, exist_ok=True)
                output_path = os.path.join(out_dir, file)

            # Formatieren
            if format_xml_file(input_path, output_path):
                count += 1
                print(f"✔️ Formatiert: {input_path}")

print(f"\n🎉 Fertig! {count} XML-Dateien formatiert.")
```
